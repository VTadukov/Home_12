# Solve problem 3

```python
ratings = pd.read_csv('ratings.csv')

group = ratings.groupby('userId').count().reset_index()
group = group[['userId', 'rating']]
group.head()
```

```python
the_one = group[group['rating']>= 100]
"""the_one - типа "избранный", как в Матрице, Нео. Здесь у нас человеки с более 100 кликов - они избранные :D"""
the_one.head()
```

```python
rating_max = ratings.groupby('userId').max().reset_index()
rating_max_timestamp = rating_max[['userId', 'timestamp']]
rating_max_timestamp.rename(columns={'timestamp': 'timestamp_max'}, inplace=True)
rating_max_timestamp.head()
```

```python
rating_min = ratings.groupby('userId').min().reset_index()
rating_min_timestamp = rating_min[['userId', 'timestamp']]
rating_min_timestamp.rename(columns={'timestamp': 'timestamp_min'}, inplace=True)
rating_min_timestamp.head()
```

```python
average = the_one.merge(rating_max_timestamp, how='inner').merge(rating_min_timestamp, how='inner')

def avg(row):
    mean_ = (row['timestamp_max'] - row['timestamp_min'])
    return mean_

average['avg'] = average.apply(avg, axis=1)
len(average)==len(the_one)
```
